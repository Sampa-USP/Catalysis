```python
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter
from qiskit.visualization import circuit_drawer

# Cria parâmetros simbólicos
theta0 = Parameter("(θ₀)")
theta1 = Parameter("(θ₁)")
theta2 = Parameter("(θ₂)")
theta3 = Parameter("(θ₃)")

# Circuito de 4 qubits
qc = QuantumCircuit(4)

# Aplicando rotações parametrizadas
qc.ry(theta0, 0)
qc.ry(theta1, 1)
qc.ry(theta2, 2)
qc.ry(theta3, 3)

# Adicionando interações CZ
qc.cz(0, 1)
qc.cz(1, 2)
qc.cz(2, 3)

# Medição
qc.measure_all()

# Desenha no estilo Matplotlib
circuit_drawer(qc, output="mpl")
```
